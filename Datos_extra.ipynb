```python
import pandas as pd
import pymysql
import numpy as np
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
# Leer datos de poblacion de CABA por comuna 

ruta='Datos/PBP_CO_1020.xls'
datos=pd.read_excel(ruta,sheet_name='PBP_CO_1020',header=2)
datos
```

```python
poblacion_por_comuna=datos.iloc[1:16,:]
poblacion_por_comuna
```

```python
poblacion_por_comuna.info()
```

```python
columns_name=[str(col)for col in poblacion_por_comuna.columns]
poblacion_por_comuna.columns=columns_name
```

```python
poblacion_por_comuna=poblacion_por_comuna.astype(int)
poblacion_por_comuna.info()
```

```python
# Utiliza melt para convertir las columnas de años en 'Año' y las poblaciones en 'Valor'
df_melted = pd.melt(poblacion_por_comuna, id_vars=['Comuna'], var_name='Fecha', value_name='Poblacion_anual')

# Agrupa por 'Año' y calcula la suma de los valores para cada año
df = df_melted.groupby('Fecha')['Poblacion_anual'].sum().reset_index()

# Imprime el DataFrame resultante
df
```

```python
# REALIZO UNA INTERPOLACION LINEAL PARA ESTIMAR LA POBLACION MENSUAL
# Convertir la columna 'año' a tipo datetime para facilitar la interpolación
df['Fecha'] = pd.to_datetime(df['Fecha'], format='%Y')

# Establecer 'año' como índice y resample a población semestral
df = df.set_index('Fecha').resample('1M').mean()

# Interpolar los valores semestrales utilizando el método de interpolación lineal
df['Poblacion_semestral'] = df['Poblacion_anual'].interpolate(method='linear')
df=df.reset_index()
df
```

```python
# Crear columns "Semestre" y columna "Año"
df['Año']=df['Fecha'].dt.year
df['Mes']=df['Fecha'].dt.month
df
```

```python
df['Poblacion_anual']=df['Poblacion_anual'].fillna(method='ffill')
df
```
